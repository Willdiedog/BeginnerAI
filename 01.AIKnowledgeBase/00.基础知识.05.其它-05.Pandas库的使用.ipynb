Pandas库的使用技巧(持续更新中)
===
使用的数据集是泰坦尼克数据集

# 1.导入

```python
import pandas as pd
data = pd.read_csv("data/Titanic.train.csv")
data.head(10)
```

# 2.Boolean Indexing
根据另一列的条件筛选当前列的值,并且只显示其中的四列

```python
data.loc[(data["Pclass"] == 3) & (data["Sex"] == "female"), ["PassengerId", "Survived", "Pclass", "Sex"]].head(10)
```

# 3.Apply Function
Apply函数是使用数据和创建新变量的常用函数之一。在对DataFrame的特定行/列应用一些函数后，它会返回相应的值。这些函数既可以是默认的，也可以是用户自定义的

```python
def num_missing(x):
    return sum(x.isnull())

data.apply(num_missing, axis=0)
```

# 4.替换缺失值
可以使用众数、平均值或者中位数来进行缺省值的补充

```python
data["Age"].fillna("23", inplace=True)
```

# 5.Pivot Table
Pandas可以用来创建MS Excel样式数据透视表

```python
pt = data.pivot_table(values=["Survived"], index=["Sex", "Pclass"])
print(pt)
```
