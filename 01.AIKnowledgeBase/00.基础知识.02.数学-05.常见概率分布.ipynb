常见概率分布

对于连续型变量，我们求概率密度函数，对于离散型变量，我门求概率分布律

# 1.离散型分布之伯努利分布-Bernouli
伯努利分布是一种离散分布,有两种可能的结果。1表示成功，出现的概率为p(其中0<p<1)。0表示失败，出现的概率为
$q=1-p$,最常见的例子就是只抛一次硬币的概率，就符合伯努利分布
$$
P_n=\begin{cases}
1-p, & n=0 \\\\
p, & n=1
\end{cases}
$$
该分布研究的是一种特殊的实验，这种实验只有两个结果要么成功要么失败，且每次实验是独立的并每次实验都有固定的成功概率p。

- 概率公式$f(x)=p^x(1-p)^{1-x},x\in(0,1)$
- 期望就是$E(X)=\sum_{i=0}^1x_if(x_i)=1*p+0*(1-p)=p$
- 方差就是$D(X)=E(x^2)-[E(x)]^2=x_1^2f(x_1^2)+x_0^2f(x_0^2)-p^2=1^2*p+0^2(1-p)-p^2=p(1-p)$

# 2.离散型分布之二项分布
二项分布就是重复n次的伯努利分布，典型的问题就是随机扔N次硬币，0次正面，1次正面，2次正面，3次正面，4次正面，
5次正面...N次正面的概率，组成二项概率分布,可以知道n=1的时候，就是伯努利分布<br/>
![images](images/00_02_05_001.png)<br/>
假设有随机变量X，表示n次进行实验次数，其中每次成功的概率为P,那么n次中有K次成功的概率，就是
$$P(X=K)=C_n^K*P^K*(1-P)^{n-K}$$
当正面的概率越小的时候，整个曲线会向左偏离，同理，如果正面的概率越大，就会向右偏离<br/>
![images](images/00_02_05_002.png)<br/>
- 期望就是$np$
- 方差$np(1-p)$

# 3.离散型分布之多项式分布
对于二项式分布和伯努利分布来说，它们都只有两个结果，要么成功，要么失败；要么是好，要么是坏；如果结果有多个，那么就变成了多项式分布，比如扔骰子
$$
P(X_1=n_1,...,X_k=n_k)=\begin{cases}
\frac{n!}{n_1!...n_k!}p_1^{n_1}...p_k^{n_k}, & \sum_{i=1}^kn_i=n \\\\
0, & otherwise
\end{cases}
$$
或者
$$
P(X_1=n_1,...,X_k=n_k)=\begin{cases}
n!\prod_{i=1}^k\frac{p_i^{n_i}}{n_i!}, & \sum_{i=1}^kn_i=n \\\\
0, & otherwise
\end{cases}
$$

# 4.离散型分布之泊松分布
当二项式分布中，n趋于无穷大的时候，二项分布就变成了泊松分布

## 4.1.泊松定理
泊松定理：设$\lambda$是一个正整数，$p=\frac{\lambda}{n}$,则有
$$\lim\limits_{n\to\infty}C_n^K*P^K*(1-P)^{n-K}=e^{-\lambda}\frac{\lambda^K}{K}$$
证明：$$\begin{eqnarray}
\because p&=&\frac{\lambda}{n}\\
&\Rightarrow& \lim\limits_{n\to\infty}P_n^k=C_n^kp^k(1-p)^{n-k}\\
&=& \lim\limits_{n\to\infty}\frac{n(n-1)...(n-k+1)}{k!}(\frac{\lambda}{n})^k(1-\frac{\lambda}{n})^{n-k}\\
&=& \lim\limits_{n\to\infty}\frac{\lambda^k}{k!}[\frac{n}{n}*\frac{n-1}{n}*\frac{n-2}{n}...*\frac{n-k+1}{n}](1-\frac{\lambda}{n})^n(1-\frac{\lambda}{n})^{-k}\\
&=& \lim\limits_{n\to\infty}\frac{\lambda^k}{k!}[1*(1-\frac{1}{n})*(1-\frac{2}{n})*...*(1-\frac{k-1}{n})](1-\frac{\lambda}{n})^n(1-\frac{\lambda}{n})^{-k}\\
&=& \lim\limits_{n\to\infty}\frac{\lambda^k}{k!}*1*(1-\frac{\lambda}{n})^n(1-\frac{\lambda}{n})^{-k}\\
&=& \lim\limits_{n\to\infty}\frac{\lambda^k}{k!}*(1-\frac{\lambda}{n})^n*1\\
&=& \frac{\lambda^k}{k!}\lim\limits_{n\to\infty}(1-\frac{\lambda}{n})^n\\
&=& \frac{\lambda^k}{k!}\lim\limits_{n\to\infty}(1-\frac{\lambda}{n})^{-\frac{n}{\lambda}(-\lambda)}\\
&=& \frac{\lambda^k}{k!}e^{-\lambda}
\end{eqnarray}$$

其中$\lambda=np$,当n>=10,p<=0.1的时候，近似效果就很好了

- 期望值：$\lambda$
- 方差：$\lambda$<br/>

在实际事例中，当一个随机事件，以固定的平均瞬时速率$\lambda$(或称密度)随机且独立地出现时，那么这个事件在单位时间(面积或体积)内存现的次数或个数就近似地服从泊松分布$P(\lambda)$
- 某一服务设施在一定时间内到达的人数
- 电话交换机接到呼叫的次数
- 汽车站台的候客人数
- 机器出现的故障数
- 自然灾害发生的次数
- 一块产品上的缺陷数
- 显微镜下单位分区内的细菌分布数
- 某放射性物质单位时间内发射出的粒子数

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10, 10), facecolor='w')
x = np.random.poisson(lam=5, size=10000)
pillar = 15
a = plt.hist(x, bins=pillar, density=True, range=[0, pillar], color='g', alpha=0.5)
plt.grid()
```

# 5.连续型分布之均匀分布
设X~U(a,b)符合在a，b间的均匀分布，则其概率密度函数为
$$f(x)=\begin{cases}
\frac{1}{b-a} & a < x < b\\\\
0 & otherwise
\end{cases}$$
- 期望$E(X)=\int_{-\infty}^{+\infty}xf(x)dx=\int_a^b\frac{1}{b-a}xdx=\frac{a+b}{2}$
- 方差$D(X)=E(X^2)-[E(X)]^2=\int_a^bx^2\frac{1}{b-a}dx-(\frac{a+b}{2})^2=\frac{(b-a)^2}{12}$<br/>

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10), facecolor='white')
x = np.random.rand(10000)
t = np.arange(len(x))
plt.hist(x, 30, color='m', alpha=0.5, label=u'EvenLy')
# plt.plot(t, x, 'r-', label=u'均匀分布')
plt.legend(loc='upper left')
plt.grid()
```

# 6.连续型分布之正态分布-高斯分布
当$\lambda$变得很大时，泊松分布就可以近似成正态分布，一般在工作中，大于20就可以近似的用正态分布来解决
$$f(X)=(\frac{1}{\sigma\sqrt{2\pi}})e^{\frac{(x-\mu)^2}{2\sigma^2}}$$
- 期望值：$\mu$
- 方差$\sigma^2$<br/>

```python
import numpy as np
import matplotlib as mpl
import math
import matplotlib.pyplot as plt

mpl.rcParams['font.sans-serif'] = [u'SimHei']  #FangSong/黑体 FangSong/KaiTi
mpl.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(12,12), facecolor='w')
mu = 0
sigma = 1
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 51)
y = np.exp(-(x - mu) ** 2 / (2 * sigma ** 2)) / (math.sqrt(2 * math.pi) * sigma)
# plt.plot(x, y, 'ro-', linewidth=2)
plt.figure(figsize=(10,10),facecolor='w')
plt.plot(x, y, 'r-', x, y, 'go', linewidth=2, markersize=8)
plt.xlabel('X', fontsize=15)
plt.ylabel('Y', fontsize=15)
plt.title(u'高斯分布函数', fontsize=18)
plt.grid(True)
```

# 7.连续型分布之指数分布
设随机变量X服从指数分布，那么其概率密度函数为
$$f(x)=\begin{cases}
\frac{1}{\theta}e^{-\frac{x}{\theta}} & x > 0\\\\
0 & x < 0
\end{cases}, \theta > 0$$
- 期望$E(X)=\int_{-\infty}^{+\infty}xf(x)dx=\int_0^{\infty}x \bullet \frac{1}{\theta}e^{-\frac{x}{\theta}}dx=\theta$
- 方差$D(X)=E(X^2)-[E(X)]^2=\int_0^{\infty}x^2 \bullet \frac{1}{\theta}e^{-\frac{x}{\theta}}dx-{\theta}^2={\theta}^2$

## 7.2.伯努利分布是指数族分布
$$
\begin{eqnarray}
a &=& b + c \tag{3}\\
  &=& d + e + f\tag{4}
\end{eqnarray}
$$

根据伯努利分布Bernouli$(\varphi)$,其中$y \in \{0, 1\}$，有如下推导：
$$\begin{eqnarray}
P\{X=x\}&=&p^y(1-p)^{1-y}\\
&=&e^{ln(p^y(1-p)^{1-y})}\\
&=&e^{lnp^y+ln(1-p)^{1-y}}\\
&=&e^{ylnp+(1-y)ln(1-p)}\\
&=&e^{ylnp+ln(1-p)-yln(1-p)}\\
&=&e^{y[lnp-ln(1-p)]+ln(1-p)}\\
&=&e^{yln\frac{p}{1-p}+ln(1-p)}
\end{eqnarray}$$
令$\eta=ln(\frac{p}{1-p})$,那么上述表达式可以写成如下形式：
$$\begin{eqnarray}
\Phi&=&e^{\eta+ln(1-\Phi)}\\
&\Rightarrow& ln\Phi=\eta+ln(1-\Phi)\\
&\Rightarrow& \frac{ln\Phi}{ln(1-\Phi)}=\eta\\
&\Rightarrow& \eta=ln\frac{\Phi}{1-\Phi}\\
&\Rightarrow& \frac{\Phi}{1-\Phi}=e^{\eta}\\
&\Rightarrow& \Phi = e^{\eta} - e^{\eta}\Phi\\
&\Rightarrow& \Phi(1+e^{\eta})=e^{\eta}\\
&\Rightarrow& \Phi=\frac{e^{\eta}}{1+e^{\eta}}\\
&\Rightarrow& \Phi=\frac{1}{1+e^{-\eta}}
\end{eqnarray}$$
推导到这里，可以看出伯努利分布属于指数族分布

# 8.连续分布之$\beta$分布
设随机变量X符合$\beta$分布，那么它的概率密度函数为$$f(x)=\begin{cases}
\frac{1}{B(\alpha,\beta)}x^{\alpha-1}(1-x)^{\beta-1}, & x \in [0,1]\\\\
0, & otherwise
\end{cases}$$
其中系数B为:$B(\alpha,\beta)=\int_0^1x^{\alpha-1}(1-x)^{\beta-1}dx=\frac{\Gamma(a)\Gamma(\beta)}{\Gamma(\alpha+\beta)}$<br/>比如我们有100个硬币，第一次实验30枚朝上，70枚朝下，我们就是它的概率为$\theta=0.3$，第二次实验29枚朝上，71枚朝
下，我们就说它的概率为$\theta=0.29$，我们做了若干次实验，那么这个概率$\theta$符合的分布，就是$\beta$分布
- 期望:$E(x)=\int_0^1x \bullet \frac{1}{B(\alpha,\beta)}x^{\alpha-1}(1-x)^{\beta-1}dx=\frac{1}{B(\alpha,\beta)}\int_0^1x^{\alpha}(1-x)^{\beta-1}dx$<br/>
因为$B(\alpha,\beta)=\int_0^1x^{\alpha-1}(1-x)^{\beta-1}dx=\frac{\Gamma(a)\Gamma(\beta)}{\Gamma(\alpha+\beta)}\Rightarrow E(x)=\frac{B(\alpha+1, \beta)}{B(\alpha, \beta)}=\frac{\frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}}{\frac{\Gamma(\alpha+\beta+1)}{\Gamma(\alpha+1)\Gamma(\beta)}}=\frac{\alpha}{\alpha+\beta}$<br/>
如果$\alpha=\beta=1$，那么这个就变成了从0到1的均匀分布

# 9.连续分布之Dirichlet分布
将$\beta$推广到高维空间，就形成了Dirichlet分布
$$f(\overrightarrow{p}|\overrightarrow{\alpha})=\begin{cases}
\frac{1}{\delta(\overrightarrow{\alpha})}\prod_{k=1}^Kp_k^{\alpha_k-1}, & p_k \in [0,1]\\\\
0, & otherwise
\end{cases}$$
简记$Dir(\overrightarrow{p}|\overrightarrow{\alpha})=\frac{1}{\delta(\overrightarrow{\alpha})}\prod_{k=1}^Kp_k^{\alpha_k-1}$，其中$\delta(\overrightarrow{\alpha})=\frac{\prod_{k=1}^K\Gamma(\alpha_k)}{\Gamma(\sum_{k=1}^K\alpha_k)}$
- 期望:$E(p_i)=\frac{\alpha_i}{\sum_{k=1}^K\alpha_k}$
